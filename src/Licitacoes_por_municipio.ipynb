```python
import pandas as pd
```

# Carregar base de dados

```python
licitacoes=pd.read_csv('../data/licitacaoes_covid_PB.csv', sep = ',')
```

# Criar dataframe com as colunas que serão utilizadas

```python
licitacoes_por_municipio = licitacoes.loc[:, ['nome_municipio', 'protocolo_licitacao', 'situacao_proposta', 'valor_proposta']]
```

```python
licitacoes_por_municipio
```

# Calcular a média do valor das licitações por município

```python
licitacoes_agrupadas_municipio = pd.DataFrame(licitacoes_por_municipio.groupby(['nome_municipio', 'protocolo_licitacao', 'situacao_proposta']).mean()['valor_proposta'])
```

```python
licitacoes_agrupadas_municipio
```

```python
licitacoes_agrupadas_municipio.to_csv('/home/thais/Downloads/licitacoes_agrupadas_municipio.csv')
```

# Criando dataframe com as licitações vencedoras

```python
licitacoes_vencedores = licitacoes[licitacoes['situacao_proposta']=='Vencedora']
```

# Calculando a soma do valor das propostas vencedoras por município

```python
pd.DataFrame(licitacoes_vencedores.groupby(['nome_municipio']).sum()['valor_proposta']).sort_values(by=['valor_proposta'], ascending=False)
```
